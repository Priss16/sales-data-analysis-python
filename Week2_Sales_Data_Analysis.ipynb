```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
np.random.seed(1)

restaurants = ['Dominos', 'McDonalds', 'KFC', 'Subway', 'Pizza Hut']
food_categories = ['Fast Food', 'Beverages', 'Desserts']
start_date = datetime(2024, 1, 1)

data = []

for i in range(50):
    order_id = i + 1001
    restaurant = np.random.choice(restaurants)
    food_category = np.random.choice(food_categories)
    quantity = np.random.randint(1, 5)
    price = np.random.randint(100, 600)
    order_date = start_date + timedelta(days=np.random.randint(0, 180))

    data.append([order_id, restaurant, food_category, quantity, price, order_date])

columns = ['Order_ID', 'Restaurant', 'Food_Category', 'Quantity', 'Price', 'Order_Date']
df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
df['Total_Sales'] = df['Quantity'] * df['Price']
df.head()
```

```python
total_revenue = df['Total_Sales'].sum()
total_orders = df['Order_ID'].nunique()

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)
```

```python
restaurant_sales = df.groupby('Restaurant')['Total_Sales'].sum().sort_values(ascending=False)
restaurant_sales
```

```python
category_sales = df.groupby('Food_Category')['Total_Sales'].sum()
category_sales
```

```python
df['Month'] = df['Order_Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Total_Sales'].sum()
monthly_sales
```

```python
print(" BUSINESS SUMMARY REPORT")
print("-" * 40)
print(f"Total Revenue Generated: ₹{total_revenue}")
print(f"Total Orders: {total_orders}")
print(f"Top Restaurant: {restaurant_sales.idxmax()}")
print(f"Best Food Category: {category_sales.idxmax()}")
print("Fast food items contributed significantly to overall revenue.")
```

```python
df.to_csv("week2_food_delivery_sales_output.csv", index=False)
print("CSV exported successfully!")
```
